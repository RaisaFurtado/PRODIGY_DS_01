```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data
df = pd.read_csv('Data.csv')

# Display the first few rows of the DataFrame
print(df.head())

# Fill missing values with the mean of the respective columns
df['Cost of Living Index'].fillna(df['Cost of Living Index'].mean(), inplace=True)
df['Rent Index'].fillna(df['Rent Index'].mean(), inplace=True)

# Group by 'Cost of Living Index' and calculate the mean 'Rent Index'
avg_Rent_Index_by_country = df.groupby('Cost of Living Index')['Rent Index'].mean()

# Select only 20 instances for plotting
avg_Rent_Index_by_country = avg_Rent_Index_by_country.head(20)

# Plot the data
avg_Rent_Index_by_country.plot(kind='bar', color='purple')
plt.title('Average Rent Index by Country')
plt.xlabel('Cost of Living Index')
plt.ylabel('Average Rent Index')
plt.xticks(rotation=45)
plt.tight_layout()

# Show the plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data
df = pd.read_csv('Data.csv')

# Display the first few rows of the DataFrame
print(df.head())

# Fill missing values with the mean of the respective columns
df['Groceries Index'].fillna(df['Groceries Index'].mean(), inplace=True)
df['Restaurant Price Index'].fillna(df['Restaurant Price Index'].mean(), inplace=True)
df['Rent Index'].fillna(df['Rent Index'].mean(), inplace=True)

# Group by 'Country' and calculate the mean 'Rent Index'
avg_Rent_Index_by_country = df.groupby('Country')['Rent Index'].mean()

# Select the top 20 countries with the highest Rent Index
top_20_highest_rent = avg_Rent_Index_by_country.sort_values(ascending=False).head(20)

# Select the top 20 countries with the lowest Rent Index
top_20_lowest_rent = avg_Rent_Index_by_country.sort_values(ascending=True).head(20)

# Plot the top 20 countries with the highest Rent Index
plt.figure(figsize=(12, 8))
top_20_highest_rent.plot(kind='bar', color='green', legend=False)
plt.title('Top 20 Countries with the Highest Rent Index')
plt.xlabel('Country')
plt.ylabel('Rent Index')
plt.xticks(rotation=45)
plt.tight_layout()

# Show the plot
plt.show()

# Plot the top 20 countries with the lowest Rent Index
plt.figure(figsize=(12, 8))
top_20_lowest_rent.plot(kind='bar', color='red', legend=False)
plt.title('Top 20 Countries with the Lowest Rent Index')
plt.xlabel('Country')
plt.ylabel('Rent Index')
plt.xticks(rotation=45)
plt.tight_layout()

# Show the plot
plt.show()
```
